```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/train.csv")
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
df.info()
```

```python
X = df.iloc[:, 0:20]
Y = df.iloc[:, 20]
```

```python
Y
```

```python
X_estandarizado = StandardScaler().fit_transform(X)
df_reescalado = pd.DataFrame(data = X_estandarizado, columns = X.columns)
df_reescalado
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_estandarizado, Y, test_size=0.3, random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB

nb_model = GaussianNB()

nb_model.fit(X_train,y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
y_pred = nb_model.predict(X_test)

print(classification_report(y_test,y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.naive_bayes import MultinomialNB

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
mnb_model = MultinomialNB()

mnb_model.fit(X_train,y_train)
```
